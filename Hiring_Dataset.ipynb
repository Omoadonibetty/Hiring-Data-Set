```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn import linear_model
from word2number import w2n
```

```python
#Connecting google colab to google drive
from google.colab import drive

#mounting the drive
drive.mount('/content/drive')
```

```python
#Reading the data and defining it as "autosales"
hiring = pd.read_csv("/content/hiring.csv")

# Displaying the content of the DataFra
hiring
```

```python
# Check for missing values in the hiring dataset and sum them up
hiring.isna().sum()
```

```python
# Calculate the median of the 'test_score(out of 10)' column in the hiring dataset
test_median = hiring['test_score(out of 10)'].median()
test_median
```

```python
# Fill missing values in the 'test_score(out of 10)' column with the calculated median value
hiring['test_score(out of 10)'] = hiring['test_score(out of 10)'].fillna(test_median)

# Display the 'test_score(out of 10)' column after filling missing values
hiring['test_score(out of 10)']
```

```python
# Accessing the column names of the 'hiring' DataFrame
hiring.columns
```

```python
# Filling missing values in the 'experience' column with 'zero'
hiring['experience'] = hiring['experience'].fillna('zero')

# Displaying the 'experience' column after filling missing values
hiring['experience']
```

```python
# Applying word_to_num function to convert text to numbers in the 'experience' column
hiring['experience'] = hiring['experience'].apply(w2n.word_to_num)

# Displaying the DataFrame after converting text to numbers in the 'experience' column
hiring
```

```python
# Extracting the 'salary($)' column from the 'hiring' DataFrame
hiring['salary($)']
```

```python
# Creating a new DataFrame 'new_hiring' by dropping the 'salary($)' column from the 'hiring' DataFrame
new_hiring = hiring.drop(['salary($)'], axis='columns')
new_hiring
```

```python
# Creating a new instance of the LinearRegression model
new_hiring_reg = linear_model.LinearRegression()
new_hiring_reg
```

```python
# Fitting the Linear Regression model to the hiring data
new_hiring_reg.fit(new_hiring, hiring['salary($)'])
```

```python
# Predicting salaries for candidates with 2 years of experience, a test score of 9, and an interview score of 6
new_hiring_reg.predict([[2, 9, 6]])
```

```python
# Predicting salaries for candidates with 12 years of experience, a test score of 10, and an interview score of 10
new_hiring_reg.predict([[12, 10, 10]])
```

```python
# Calculating the R^2 score of the linear regression model
new_hiring_reg.score(new_hiring, hiring['salary($)'])
```
